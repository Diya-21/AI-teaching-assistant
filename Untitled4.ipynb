```python
import re, pandas as pd
import pdfplumber
from dateutil import parser as dateutil_parser

def iso_date_from_text(text):
    try:
        import re
        month_names = r"(Jan|January|Feb|February|Mar|March|Apr|April|May|Jun|June|Jul|July|Aug|August|Sep|Sept|September|Oct|October|Nov|November|Dec|December)"
        m = re.search(rf"(\d{{1,2}}\s+{month_names}\s+\d{{4}}|\b{month_names}\s+\d{{1,2}},?\s+\d{{4}}|\d{{4}}-\d{{2}}-\d{{2}})", text, re.I)
        if m:
            dt = dateutil_parser.parse(m.group(0), dayfirst=False)
            return dt.date().isoformat()
    except:
        return None
    return None

def extract_salaries(text):
    return re.findall(r"[\d{1,3},]+(?:\d{3})+\s*(INR|USD|EUR)?|\d+\s*(INR|USD|EUR)", text)

def skill_ratings(text):
    return re.findall(r"([A-Za-z\+\# ]{2,30})\s+(?:expertise|proficiency|score|scores?)\s+(?:at\s+)?(?:a\s+)?(\d(?:\.\d)?)(?:\s*out of\s*(\d+))?", text, re.I)

def extract_pairs_from_text(full_text):
    paragraphs = [p.strip() for p in re.split(r"\n\s*\n", full_text) if p.strip()]
    rows = []
    for p in paragraphs:
        if re.search(r"\bborn\b", p, re.I):
            dob = iso_date_from_text(p)
            if dob:
                rows.append({"Key":"Date of birth","Value":dob,"Comments":p})
            nm = re.match(r"([A-Z][a-z]+(?:\s+[A-Z][a-z]+)+)\s+was born", p)
            if nm:
                rows.append({"Key":"Name","Value":nm.group(1).strip(),"Comments":p})
        sal = extract_salaries(p)
        if sal:
            rows.append({"Key":"Salary","Value":str(sal),"Comments":p})
        perc = re.findall(r"\b\d{1,3}\.\d+%|\b\d{1,3}%|\b\d{1,3}\.\d+\b\s*%?", p)
        for pc in perc:
            rows.append({"Key":"Percentage/Score","Value":pc,"Comments":p})
        for sk in skill_ratings(p):
            rows.append({"Key":"Skill Rating","Value":f"{sk[0].strip()} — {sk[1]}", "Comments":p})
        # fallback
        rows.append({"Key":"Paragraph","Value":p[:200],"Comments":p})
    # remove duplicates
    seen = set()
    final=[]
    for r in rows:
        kv = (r["Key"], r["Value"])
        if kv in seen: continue
        seen.add(kv)
        final.append(r)
    return final

pdf_path = "assignment_files/Data Input.pdf"
with pdfplumber.open(pdf_path) as pdf:
    full_text = "\n\n".join(page.extract_text() or "" for page in pdf.pages)

pairs = extract_pairs_from_text(full_text)
df = pd.DataFrame(pairs)
df.to_excel("Generated_Output.xlsx", index=False)
print("Saved Generated_Output.xlsx with", len(df), "rows")
```
